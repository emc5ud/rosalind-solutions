## Problem

Given: A collection of DNA strings in FASTA format having total length at most 10 kbp.

Return: The adjacency list corresponding to O3. You may return edges in any order.

```python
# Sample Dataset
sample = '''>Rosalind_0498
AAATAAA
>Rosalind_2391
AAATTTT
>Rosalind_2323
TTTTCCC
>Rosalind_0442
AAATCCC
>Rosalind_5013
GGGTGGG
'''.split()
```

```python
def fasta_read(lines):
    strings = {}
    i = 0
    while i < len(lines):
        key = lines[i][1:].strip()
        i += 1
        string = ''
        while i < len(lines) and lines[i][0] !=  '>':
            string += lines[i].strip()
            i += 1
        strings[key] = string
    return strings


def solution(s): 
    print(s.count('A'), s.count('C'), s.count('G'), s.count('T'))
```

```python
# Sample  Output
def solution(s):
    strings = fasta_read(s)
    nodes = []
    keys = list(strings.keys())
    for s_key, s in strings.items():
        for t_key, t in strings.items():
            if s == t:
                continue
            if s[-3:] == t[:3]:
                print(s_key, t_key)
```
